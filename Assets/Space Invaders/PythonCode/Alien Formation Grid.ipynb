```python
import matplotlib.pyplot as plt
import numpy as np

# Alien formation parameters
rows = 5
columns = 11
spacing = 2.0

# Calculate the total width and height of the grid
total_width = (columns - 1) * spacing
total_height = (rows - 1) * spacing

# Calculate offsets to center the formation
offset_x = -total_width / 2.0
offset_y = -total_height / 2.0

# Create a figure and axis
fig, ax = plt.subplots()

# Draw the grid
for row in range(rows):
    for col in range(columns):
        x = offset_x + col * spacing
        y = offset_y + row * spacing
        ax.plot(x, y, 'bo')  # plot alien position as a blue dot

# Draw the horizontal and vertical lines for spacing and offsets
ax.arrow(offset_x, 0, total_width, 0, head_width=0.3, head_length=0.5, fc='gray', ec='gray', linestyle='--')
ax.arrow(0, offset_y, 0, total_height, head_width=0.3, head_length=0.5, fc='gray', ec='gray', linestyle='--')

# Label spacing between columns and rows
for col in range(columns - 1):
    x = offset_x + col * spacing + spacing / 2
    ax.text(x, offset_y - spacing / 2, f'{spacing}', ha='center')

for row in range(rows - 1):
    y = offset_y + row * spacing + spacing / 2
    ax.text(offset_x - spacing / 2, y, f'{spacing}', va='center')

# Label total width, height, and offsets
ax.text(0, offset_y - spacing, 'Total Width', ha='center', fontsize=10, color='red')
ax.text(offset_x - spacing, 0, 'Total Height', va='center', rotation=90, fontsize=10, color='red')

ax.text(offset_x, 0, 'offsetX', ha='right', color='green')
ax.text(0, offset_y, 'offsetY', va='top', color='green')

# Add grid labels
ax.text(0, offset_y + total_height + spacing / 2, 'Y-Axis', ha='center')
ax.text(offset_x + total_width + spacing / 2, 0, 'X-Axis', va='center')

# Set aspect ratio, grid and limits
ax.set_aspect('equal')
ax.grid(True)
ax.set_xlim([offset_x - spacing, offset_x + total_width + spacing])
ax.set_ylim([offset_y - spacing, offset_y + total_height + spacing])

# Show the plot
plt.title('Alien Formation Grid with Spacing and Offsets')
plt.show()
```
